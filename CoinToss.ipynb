```python
from random import random

def fair_toss() :
    r = random()
    if r > 0.5 :
        return 0
    else :
        return 1
fair_toss()
```

```python
def testcase(func, times) :
    result = {}
    for i in range(times) :
        f = func()
        if f in result :
            result[f] += 1
        else :
            result[f] = 1
    return result
```

```python
testcase(fair_toss,10000)
```

```python
def testcase_withvariable(func,attr,times) : 
    result = {}
    for i in range(times) :
        f = func(attr)
        if f in result :
            result[f] += 1
        else :
            result[f] = 1
    return result
```

```python
def dice() :
    d = 6
    r = random()
    if r < 1/d :
        return 1
    elif r < 2/d :
        return 2
    elif r < 3/d :
        return 3
    elif r < 4/d :
        return 4
    elif r < 5/d :
        return 5
    else:
        return 6
```

```python
testcase(dice,10000)
```

```python
def dice_new() :
    d = 6
    return int(6*random())
dice_new()
```

```python
testcase(dice_new,1000)
```

```python
def dice_n(n) :
    d = 1/n
    prob = []
    r = random()
    
    for i in range(n) :
        prob.append((i+1)*d)
    print(prob)
    for i,p in enumerate(prob) :
        
        if r < p :
            return i
dice_n(6)
        
```

```python
testcase_withvariable(dice_n,6,10000)
```

weighted probalities 

```python
def weighted_probability(prob) :
    r = random()
    for i in range(len(prob)) :
        if r < prob[i] :
            return i
    return (len(prob) - 1)
weighted_probability([0.3,0.7])
```

```python
testcase_withvariable(weighted_probability,[0.1,0.4,0.5],100)
```
