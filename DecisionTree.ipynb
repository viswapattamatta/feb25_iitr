```python
class DecisionTree() :
    criteria = None
    boundary = None
    left = None
    right = None
    
    def __init__(self,criteria,boundary,left,right) :
        self.criteria =  criteria
        self.boundary = boundary
        self.left = left
        self.right = right
        
    def answer(self,circumstance) :
        if(circumstance[self.criteria] < self.boundary) :
            return self.left.answer(circumstance)        
        else :
            return self.right.answer(circumstance)
        
class yes(DecisionTree) :
    def __init__(self):
        pass
    def answer(self, circumstance) :
        return True

class no(DecisionTree) :
    def __init__(self) :
        pass
    def answer(self,circumstance) :
        return False
```

```python
distance = DecisionTree("distance",5,yes(),no())
learning = DecisionTree("learning",2,no(),distance)
salary = DecisionTree("salary",50,no(),learning)

salary.answer({
    "salary" : 51,
    "learning" : 2,
    "distance" : 5
})
```
