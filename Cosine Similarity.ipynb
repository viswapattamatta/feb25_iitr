#### Unit vector

```python
import math

def unitVector(arr) :
    t = 0
    uv = []
    for i in range(len(arr)) :
        t += arr[i]**2
    d = math.sqrt(t)
    
    for j in range(len(arr)) :
        uv.append(arr[j]/d)
    return uv

unitVector([2,3])    
    
```

#### cosine similarity 
cos(a-b) = cosa cosb + sina sinb; 
x1x2+y1y2

```python
def cos_similarity(arr1,arr2) :
    result = 0
    
    arr1 = unitVector(arr1)
    arr2 = unitVector(arr2)
    print(f"unit vectors are -> {arr1} and {arr2}")
    for i in range(len(arr1)) :
        result += arr1[i]*arr2[i]
    return result
```
