```python
# one variable x

def f(x) : 
    y = 10* (x**2) - 35*x + 2
    return  y

def minimizer(f,epouchs=100,lr=0.01,start=0.1) : 
    x = start
    d = 0.00000001
    rateOfChange = 0
    for i in range(epouchs) : 
        rateOfChange = (f(x+d) - f(x))/d
        x = x - lr*rateOfChange
        print(f"Change is {rateOfChange} and new x is {x}")
    return x

#other option tried
#rateOfChange = (f(x+lr) - f(x))
#        x = x - x*rateOfChange
```

```python
# minimize 2 variables

def minimizer2(f,epouchs=100,lr=0.01,start=0.5) :
    x1 = start
    x2 = start
    d = 0.00000001
    print(f"value at the start {f(x1,x2)}")
    
    for i in range(epouchs) :
        changex1 = (f(x1+d,x2) - f(x1,x2) ) / d
        changex2 = (f(x1,x2+d) - f(x1,x2)) / d
        x1 = x1 - changex1 * lr
        x2 = x2 - changex2 * lr
        print(f"new x1 - {x1}, new x2 - {x2} - value {f(x1,x2)}")
    return (x1,x2)
```

```python
def p(x1, x2): 
    return (x1-x2)**2 + (x1-10)**2 + (x2 - 4)**2 + 10
```

```python
minimizer2(p)
```
