```python
class TwoPointsLine() :
    ''' This is class for 2 points linear regression'''
    
    def ___init___(self) :
        print(f"Initialzing twopointsline..")
    
    def fit(self,x,y) :
        x1,y1 = x
        x2,y2 = y
        m = (y2-y1)/(x2-x1)
        c = y2 - m*x2
        self.m = m
        self.c = c
        print(f"m is {self.m}, C is {self.c}")
    
    def predict(self,x) :
        predicted_y = []
        for i in x :
            predicted_y.append(self.m * i + self.c)
        return predicted_y
            
```

```python
model = TwoPointsLine()
model.fit([2,3],[5,9])
model.predict([2,3,-1])
```
