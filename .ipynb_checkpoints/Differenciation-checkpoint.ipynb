## Differenciation

```python
def myFunc(arr) :
    x , y = arr
    print(f"x is {x}, y is {y}")
    return (x**2 + 2*x + 4*y)
    
myFunc([2,3])
```

```python
def differenciation(f,arr) :
    x , y = arr 
    func = f(arr)
    d = 1e-8
    x2 = x + d
    print (x2)
    func2 = f([x2,y])
    print(func2)

differenciation(myFunc,[2,3])
```
