```python

def find_variable (arr, var) :
    length = len(arr)
    result = []
    if(len(var) == 0) :
        return (arr[1]/arr[0])
    else:
        temp = 0
        for i in range(0,len(var)) :
            temp += arr[i]*var[i]
    variable = (arr[-1] - temp) / arr[-2]    
    return variable
        
        
def divide(arr,d) :
    for i in range(len(arr)) :
        arr[i] = arr[i]/abs(d)
    return arr   

def add_and_eliminate(arr1, arr2) :
    result = []
    for i in range(len(arr1)) :
        temp = arr1[i]+arr2[i]
        if(i != len(arr1)-2) : result.append(temp)      
    return result   

def sub_and_eliminate(arr1, arr2) :
    result = []
    for i in range(len(arr1)) :
        temp = arr1[i] - arr2[i]
        if(i != len(arr1)-2) : result.append(temp)      
    return result 


def eliminate_var(arr) : 
   
    eqs = len(arr)
    e_position = len(arr[0]) - 2
    i = 0
    result = []

    #normalize 
    for j in range(0,eqs) :
        arr[j] = divide(arr[j],arr[j][e_position])
        
    while i < eqs - 1 :       
        
        if (arr[i][e_position] < 0 and arr[i+1][e_position] < 0) : 
            result.append (sub_and_eliminate(arr[i],arr[i+1]) )
            
        elif ( (arr[i][e_position]) < 0 or (arr[i+1][e_position]) < 0 ):
            #add
             result.append( add_and_eliminate(arr[i],arr[i+1]) )
        else :
            result.append (sub_and_eliminate(arr[i],arr[i+1]) )
        i+=1    
    return result
      
        
#eliminate_var([[4, 2, 1, 5],[1, -3, -1, 3],[3, 2, 2, 10]])     
#eliminate_var([[5.0, -1.0, 8.0], [2.5, -2.0, 8.0]])
#eliminate_var([[1,1,1,1,10], [0,2,-1,2,5], [0,-1,3,-4,-1], [3,2,1,-1,12]])
```

```python
def equations(arr) :
    result = arr
    temp = [arr[0]]
    for i in range(len(arr)) :
        if (len(result) > 1) :
            print(f"elimitate variable from - {result}")
            result = eliminate_var(result)
            temp.append(result[0])
           
    temp = temp[::-1]
    print(f"Substitute to find value of variables in - {temp}")
    variables = []
    for k in range(0,len(temp)) : 
        variables.append(find_variable(temp[k],variables))
            
    return variables

#Tests
#equations([[4, 2, 1, 5],[1, -3, -1, 3],[3, 2, 2, 10]])       
#equations([[1,-2,8],[2,1,5]])

#equations([[3,-1,2,5], [4,2,-1,6], [5,-3,1,1]])

#equations([[1,1,1,1,10], [0,2,-1,2,5], [0,-1,3,-4,-1], [3,2,1,-1,12]])

equations([[2,5,-9,3,151],[5,6,-4,2,103],[3,-4,2,7,16],[11,7,4,-8,-32]])
```
