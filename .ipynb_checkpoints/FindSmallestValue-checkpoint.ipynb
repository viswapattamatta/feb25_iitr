```python
def findSmallest(arr) :
    first = 0
    last = len(arr)
    while True :
        
        mid = (first + last) // 2
        if arr[mid] < arr[mid + 1] and arr[mid] < arr[mid -1] :
            return arr[mid]
        elif arr[mid] > arr[mid + 1] : # .. move right
            first = mid
        else : # if mid < mid + 1 ... move left
            last = mid
        
#findSmallest([10, 9, 8, 6.5, 4.1, 3.2, 2, 4, 4.5, 6])
#findSmallest([150,100,94,54,36,31,28,17,15,20,21,29,97,150])
findSmallest([9,7,6,4,0,-2,1])
```

```python
def function(x) :
    y = 10* (x**2) - 35*x + 2
    return y

def findx(f) :
    delta = 10**-4
    x = 0
    y = f(x)
    print(f"y at 0 is {y}")
    inc = 1
    
    while True :
        if f(x+inc) < y :
            x += inc
            y = f(x)
            print(f"x is {x}, y is {y}")
        elif f(x-inc) < y :
            x -= inc
            y = f(x)
            print(f"x is {x}, y is {y}")
        else :
            if inc == 1:
                inc = 10**-4
            else:
                break            
    return x
```

```python
findx(function)
```

```python
#Generic code for any convex equation

def findMin(eq) :
    return p(eq,1.7500000000000275)
   
def p(eq,x):
    y = 0
    power = len(eq)-1
    for i in range(len(eq)) :
        y += eq[i]* x**(power-i)
        print(f"i is {i} eq is {eq[i]}, power is {power-i} and y is {y}")
    print(f"final y is {y}")
    return y
    
    
findMin([10,-35,2])
    
```
