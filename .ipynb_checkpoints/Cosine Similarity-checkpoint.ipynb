Unit vector

```python
import math

def unitVector(arr) :
    t = 0
    uv = []
    for i in range(len(arr)) :
        t += arr[i]**2
    d = math.sqrt(t)
    
    for j in range(len(arr)) :
        uv.append(arr[j]/d)
    return uv

unitVector([2,3])    
    
```
