## Differenciation

```python
def myFunc(arr) :
    x , y = arr
    return (x**2 + 2*x + 4*y)
    
myFunc([2,3])
```

```python
def differenciation(f,arr,index) :
    z1 = f(arr)
    d = 1e-8
    arr[index] = arr[index] + d
    z2 = f(arr)
    diff = (z2-z1) / d
    return diff

differenciation(myFunc,[2,3],0)
```
